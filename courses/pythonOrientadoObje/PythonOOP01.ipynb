## Operadores de comparacin

En esta seccin se revisarn los operadores de comparacin en Python.

```python
a = 10
b = 5

# Igual que
print(a == b)

# No es igual
print(a != b)

# Mayor que
print(a > b)

# Mayor o igual
print(a >= b)

# Menor que
print(a < b)

# Menor o igual
print(a <= b)

print()

# Conjuncin
print(True and False)
print(True and True)

print()

# Disyuncin
print(True or False)
print(False or False)
```

## Ciclos

En esta seccin se vern los ciclos for y while. En Python no existen los ciclos do while.

### Ciclo 'for'

Este ciclo sirve para recorrer una lista de elementos. Estos ciclos son recomendables para cuando se sabe el nmero de veces que se va a iterar.

```python
nombres = ["Perla", "Jos", "Ximena", "Jess"]

# Se usa la variable de control 'nombre' y se itera por la lista de nombres
for nombre in nombres:
    # Aqu se puede hacer cualquier operacin con la variable de control
    # Es importante tomar en cuenta la identacin (espacios) para denotar que el print va dentro del for
    print(nombre)
    
print()

# Con la funcin 'enumerate' se pueden obtener los ndices de la lista as como los elementos
for index, nombre in enumerate(nombres):
    print(index, nombre)
    
# Con la funcin 'range' se puede crear un rango y se puede usar con un for
for num in range(1, 100):
    print(num, end=' ')
```

### Ciclos 'while'

Este ciclo se ejecuta mientras se cumpla una condicin. Estos ciclos son recomendables cuando no se sabe cuntas veces se va a iterar

```python
# Usar un ciclo para pedirle una contrasea a un usuario
password = '1234'
input_password = ''

while password != input_password:
    input_password = input('Ingresa la contrasea: ')
    
    # Se puede usar la sentencia 'break para salir de un chiclo'
    if input_password == 'unam':
        break
```

## Funciones

En Python las funciones se declaran con la palabra reservada 'def', como se puede ver en el siguiente bloque de cdigo

```python
# Definicin bsica de una funcin
def saludar():
    print("Hola!")
    
# De esta forma se llama una funcin
saludar()
```

Entre los parntesis de la declaracin de una funcin van los argumentos, de la siguiente manera

```python
def saludar(nombre):
    print("Hola", nombre, end="!\n")
    
saludar("Perla")
```

De esta forma se declara una funcin con parmetros que pueden tener un valor ya establecido

```python
def saludar(nombre="Ivn"):
    print("Hola", nombre, end="!\n")
    
# Llamar a la funcin con el parmetro 'nombre'
saludar("Jess")

# Llamar a la fncin sin el parmetro 'nombre'
saludar()
```

En el siguiente bloque de cdigo se ver una funcin recursiva en Python usando 'return'

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

resultado = factorial(5)
print(resultado)
```
